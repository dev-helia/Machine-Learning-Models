```python
import numpy as np
import matplotlib.pyplot as plt

# Define convex and concave functions and gradients
def f_conv(x): return x**2
def grad_conv(x): return 2*x

def f_conc(x): return -x**2
def grad_conc(x): return -2*x

# Gradient descent settings
lr = 0.1
steps = 20
x0 = 5

# Run gradient descent for convex function
x = x0
history_conv = [x]
for _ in range(steps):
    x -= lr * grad_conv(x)
    history_conv.append(x)

# Plot convex function and descent path
x_vals = np.linspace(-6, 6, 400)
plt.plot(x_vals, f_conv(x_vals))
plt.scatter(history_conv, [f_conv(x) for x in history_conv])
plt.title("Gradient Descent on Convex Function f(x)=x^2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()

# Run gradient descent for concave function
x = x0
history_conc = [x]
for _ in range(steps):
    x -= lr * grad_conc(x)
    history_conc.append(x)

# Plot concave function and descent path
plt.plot(x_vals, f_conc(x_vals))
plt.scatter(history_conc, [f_conc(x) for x in history_conc])
plt.title("Gradient Descent on Concave Function f(x)=-x^2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#  f(x) = -x^2 
def f(x): return -x**2
def grad(x): return -2*x

# 
lr = 0.1
steps = 20
x = 5.0
history = [x]

for _ in range(steps):
    x = x + lr * grad(x)   #   
    history.append(x)

# 
x_vals = np.linspace(-6, 6, 400)
plt.plot(x_vals, f(x_vals))
plt.scatter(history, [f(x) for x in history])
plt.title("Gradient Ascent on Concave Function f(x)=-x^2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```
