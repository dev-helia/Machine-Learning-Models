```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# 生成示例数据
np.random.seed(42)
x1 = np.random.randint(1000, 3000, 100)  # 比如房屋面积（范围大）
x2 = np.random.randint(1, 5, 100)        # 比如卧室数量（范围小）

X = np.vstack((x1, x2)).T

# Scaling（缩放到 0~1）
scaler_minmax = MinMaxScaler()
X_scaled = scaler_minmax.fit_transform(X)

# Normalization（标准化）
scaler_std = StandardScaler()
X_normalized = scaler_std.fit_transform(X)

# 可视化对比
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# 原始数据
axes[0].set_title("🌟原始数据")
sns.scatterplot(x=X[:, 0], y=X[:, 1], ax=axes[0])
axes[0].set_xlabel("房屋面积")
axes[0].set_ylabel("卧室数量")

# 缩放后
axes[1].set_title("📏Scaling (Min-Max)")
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], ax=axes[1])
axes[1].set_xlabel("面积_scaled")
axes[1].set_ylabel("卧室_scaled")

# 标准化后
axes[2].set_title("📐Normalization (Standard)")
sns.scatterplot(x=X_normalized[:, 0], y=X_normalized[:, 1], ax=axes[2])
axes[2].set_xlabel("面积_std")
axes[2].set_ylabel("卧室_std")

plt.tight_layout()
plt.show()
```
