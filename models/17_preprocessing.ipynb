```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# 
np.random.seed(42)
x1 = np.random.randint(1000, 3000, 100)  # 
x2 = np.random.randint(1, 5, 100)        # 

X = np.vstack((x1, x2)).T

# Scaling 0~1
scaler_minmax = MinMaxScaler()
X_scaled = scaler_minmax.fit_transform(X)

# Normalization
scaler_std = StandardScaler()
X_normalized = scaler_std.fit_transform(X)

# 
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# 
axes[0].set_title("")
sns.scatterplot(x=X[:, 0], y=X[:, 1], ax=axes[0])
axes[0].set_xlabel("")
axes[0].set_ylabel("")

# 
axes[1].set_title("Scaling (Min-Max)")
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], ax=axes[1])
axes[1].set_xlabel("_scaled")
axes[1].set_ylabel("_scaled")

# 
axes[2].set_title("Normalization (Standard)")
sns.scatterplot(x=X_normalized[:, 0], y=X_normalized[:, 1], ax=axes[2])
axes[2].set_xlabel("_std")
axes[2].set_ylabel("_std")

plt.tight_layout()
plt.show()
```
