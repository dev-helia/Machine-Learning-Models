线性回归的一种方法

用正规方程法（Normal Equation）直接算出 θ 参数

```python
import numpy as np

# 特征矩阵 X，加一列全 1（x₀）
X = np.array([
    [1, 2104, 5, 1, 45],
    [1, 1416, 3, 2, 40],
    [1, 1534, 3, 2, 30],
    [1,  852, 2, 1, 36]
])

# 标签 y：房价（单位 $1000）
y = np.array([460, 232, 315, 178])
```

```python
# 公式：θ = (XᵗX)⁻¹ Xᵗy
theta = np.linalg.inv(X.T @ X) @ X.T @ y

print("正规方程求得的 θ 参数为：")
print(theta)
```

```python
x_new = np.array([1, 2000, 3, 2, 40])
predicted_price = x_new @ theta
print(f"预测房价为：{predicted_price:.2f} 千美元")
```
